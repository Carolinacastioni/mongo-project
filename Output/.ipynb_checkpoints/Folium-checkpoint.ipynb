## Import the final DataFrame

```python
import pandas as pd
from pandas.io.json import json_normalize
import folium
import numpy as np
```

```python
df_folium = pd.read_csv("../Output/df_tableau.csv")
df_folium.head()
```

## New office coordinates - Folium visualization

```python
new_office = folium.Map(location=[40.757929, -73.985506], zoom_start=15)

folium.Marker([40.757929, -73.985506], icon=folium.Icon(color='red', icon='cloud')).add_to(new_office)

new_office
```

## Starbucks = Blue

```python
starbucks = df_folium[df_folium.name == 'Starbucks']
```

```python
starbucks = starbucks.drop(columns=['Unnamed: 0', 'name'])
```

```python
s = starbucks.values.tolist()
```

```python
for coord in s:
    folium.Marker(
    location=(coord),
    popup='Starbucks',
    icon=folium.Icon(color='lightblue', icon='cloud')).add_to(new_office)
```

```python
design = df_folium[df_folium.name == 'Design companies']
design = design.drop(columns=['Unnamed: 0', 'name'])

d = design.values.tolist()
```
